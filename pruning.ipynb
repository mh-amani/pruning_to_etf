# pruning a two-layer net trained on mnist

```python
from utils_tf import load_mnist_evod
x_train, y_train, x_test, y_test = load_mnist_evod()
```

```python
x_train.shape
```

```python
from utils_tf import neural_net, kernel_of_nn, train_teacher_student_gaussian_input
from tensorflow.keras import backend as K
bce = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
d = 784
Ns = [1000]
Ms = [5]
out_dim = 2
num_iter = 5000
bsize = 3000

target_model = neural_net(d, Ns, out_dim)
target_model.compile(optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),)

target_model.fit(
    x_train, y_train, batch_size=bsize,
    epochs=20,
    validation_data=(x_test, y_test),
)

```

```python
def pred(model):
  return np.argmax(model(x_test), axis=-1)

def ret_acc(model):
  # print("test_accuracy = ", np.mean(y_test_evod * 1.0 == np.ndarray.flatten((1 + np.sign(model(x_test)))/2)))
  return( np.mean(y_test * 1.0 == pred(model)))

def ret_mse(model, x_test):
  # Using 'auto'/'sum_over_batch_size' reduction type.
  mse = tf.keras.losses.MeanSquaredError()
  return(mse(y_test * 1.0, pred(model, x_test)).numpy())

ret_acc(target_model)
```

```python
pruned_model = neural_net(d, Ms, out_dim)

optimizer = tf.keras.optimizers.Adam()
train_teacher_student_gaussian_input(pruned_model, target_model, num_iter, bsize, optimizer)
print(np.round(kernel_of_nn(pruned_model), 2))
```
